```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
%matplotlib inline
```

```python
model = LinearRegression(normalize=True)
```

### Creating Data

```python
x = np.random.rand(200,1)*20
y = x**4 - 2 * x**3 - 3 * x**2 - 4 * x - 5
y = y + np.random.normal(scale=4000,size=y.shape)
y = y.reshape(-1, 1)
x = x.reshape(-1, 1)

grade=[1,2,4,8,16]
trainerror=[]
testerror=[]
```

### Split Data And Normalize

```python
n = int(x.shape[0] * 0.8)
xtrain = x[:n].reshape(-1, 1)
ytrain = y[:n].reshape(-1, 1)
xtest = x[n:].reshape(-1, 1)
ytest = y[n:].reshape(-1, 1)
```

## Question 1

```python
#1
model.fit(xtrain,ytrain)
xline = np.arange(0,20, 0.1)
xline = xline.reshape(-1, 1)

plt.figure()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("K = 1")
plt.plot(xline,model.predict(xline),color='#9a40f5',label="Model")
plt.plot(xtrain,ytrain,'bo',label="Train")
plt.plot(xtest,ytest,'o',color='#33caba',label="Test")
plt.legend()

trainerror.append(np.mean((ytrain - model.predict(xtrain)) ** 2) / 2)
testerror.append(np.mean((ytest - model.predict(xtest)) ** 2) / 2)

train_sizes , train_score, test_score = learning_curve(model,xtrain,ytrain, cv=5)
plt.figure()
plt.xlabel("Train Size")
plt.ylabel("Score")
plt.title("Learning Curve : K = 1")
plt.plot(train_sizes, train_score.mean(axis=1),'o-',label="Train")
plt.plot(train_sizes, test_score.mean(axis=1),'o-',label="Test")
plt.legend()
```
